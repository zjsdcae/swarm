```python
# header to start
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt

import scipy.linalg
import pickle

import warnings

warnings.filterwarnings('ignore')
```

```python
def getLaplacian(E, n_vertices, directed):
    A = np.zeros([n_vertices,n_vertices]) # Adjacency
    D = np.zeros([n_vertices,n_vertices]) # Degree
    # undirected graph the graph is symmetric           
    if directed == False:    
        for edge in E:
            A[edge[0],edge[1]] = A[edge[1],edge[0]] = 1
    # for directed graph 
    else:
        for edge in E:
            A[edge[1],edge[0]] = 1
    # the degree matrix equals the sum of the row
    for degree in range(n_vertices):
        for adj in A[degree]:
            D[degree][degree] += adj
    # Laplacian = Diag - Adjacency  
    L = D - A
    return L
```

## Question 1
For a digraph D, the node i has a directed path to node j only if the e^(-L(D))[i][j] is greater than 0. In this question, we caculate the exponential of Laplacian matrix of diagraph and check whether the value is greater than 0.

The minimum path is caculated using the Floyd algorithm, the path of the two nodes could be caculated testing whether there is a node which can reduce the length of the two nodes.

```python
# Question 1

inf = 999
def getMinPath(E, n_vertices):
    if [0,9] in E:
        return 1
    map = [[inf]*n_vertices for i in range(n_vertices)]
    for i in range(n_vertices):
        map[i][i] = 0
    for path in E:
        map[path[0]][path[1]] = 1
    for k in range(n_vertices):
        # start
        for i in range(n_vertices):
            # end
            for j in range(n_vertices):
                map[i][j] = min(map[i][j], map[i][k] + map[k][j])
    return map[0][9]

# Graph10 a
with open('Exercise1_Graph10a/graph1.pickle', 'rb') as f:
    E1 = pickle.load(f)
    L1 = getLaplacian(E1, 10, directed=True)
    directed = scipy.linalg.expm(-L1)[9][0]
    print("unconnected") if directed == 0 else print("connected, the minimum path is ",getMinPath(E1,10))

with open('Exercise1_Graph10a/graph2.pickle', 'rb') as f:
    E1 = pickle.load(f)
    L1 = getLaplacian(E1, 10, directed=True)
    directed = scipy.linalg.expm(-L1)[9][0]
    print("unconnected") if directed == 0 else print("connected, the minimum path is ",getMinPath(E1,10))
    
with open('Exercise1_Graph10a/graph3.pickle', 'rb') as f:
    E1 = pickle.load(f)
    L1 = getLaplacian(E1, 10, directed=True)
    directed = scipy.linalg.expm(-L1)[9][0]
    print("unconnected") if directed == 0 else print("connected, the minimum path is ",getMinPath(E1,10))
    
with open('Exercise1_Graph10a/graph4.pickle', 'rb') as f:
    E1 = pickle.load(f)
    L1 = getLaplacian(E1, 10, directed=True)
    directed = scipy.linalg.expm(-L1)[9][0]
    print("unconnected") if directed == 0 else print("connected, the minimum path is ",getMinPath(E1,10))
print('\n')

# Graph10b 
with open('Exercise1_Graph10b/graph1.pickle', 'rb') as f:
    E1 = pickle.load(f)
    L1 = getLaplacian(E1, 10, directed=True)
    directed = scipy.linalg.expm(-L1)[9][0]
    print("unconnected") if directed == 0 else print("connected, the minimum path is ",getMinPath(E1,10))

with open('Exercise1_Graph10b/graph2.pickle', 'rb') as f:
    E1 = pickle.load(f)
    L1 = getLaplacian(E1, 10, directed=True)
    directed = scipy.linalg.expm(-L1)[9][0]
    print("unconnected") if directed == 0 else print("connected, the minimum path is ",getMinPath(E1,10))
    
with open('Exercise1_Graph10b/graph3.pickle', 'rb') as f:
    E1 = pickle.load(f)
    L1 = getLaplacian(E1, 10, directed=True)
    directed = scipy.linalg.expm(-L1)[9][0]
    print("unconnected") if directed == 0 else print("connected, the minimum path is ",getMinPath(E1,10))
    
with open('Exercise1_Graph10b/graph4.pickle', 'rb') as f:
    E1 = pickle.load(f)
    L1 = getLaplacian(E1, 10, directed=True)
    directed = scipy.linalg.expm(-L1)[9][0]
    print("unconnected") if directed == 0 else print("connected, the minimum path is ",getMinPath(E1,10))
print('\n')    

#Graph100 a
with open('Exercise1_Graph100a/graph1.pickle', 'rb') as f:
    E1 = pickle.load(f)
    L1 = getLaplacian(E1, 100, directed=True)
    directed = scipy.linalg.expm(-L1)[9][0]
    print("unconnected") if directed == 0 else print("connected, the minimum path is ",getMinPath(E1,100))

with open('Exercise1_Graph100a/graph2.pickle', 'rb') as f:
    E1 = pickle.load(f)
    L1 = getLaplacian(E1, 100, directed=True)
    directed = scipy.linalg.expm(-L1)[9][0]
    print("unconnected") if directed == 0 else print("connected, the minimum path is ",getMinPath(E1,100))
    
with open('Exercise1_Graph100a/graph3.pickle', 'rb') as f:
    E1 = pickle.load(f)
    L1 = getLaplacian(E1, 100, directed=True)
    directed = scipy.linalg.expm(-L1)[9][0]
    print("unconnected") if directed == 0 else print("connected, the minimum path is ",getMinPath(E1,100))
    
with open('Exercise1_Graph100a/graph4.pickle', 'rb') as f:
    E1 = pickle.load(f)
    L1 = getLaplacian(E1, 100, directed=True)
    directed = scipy.linalg.expm(-L1)[9][0]
    print("unconnected") if directed == 0 else print("connected, the minimum path is ",getMinPath(E1,100))
print('\n')

#Graph100 b
with open('Exercise1_Graph100b/graph1.pickle', 'rb') as f:
    E1 = pickle.load(f)
    L1 = getLaplacian(E1, 100, directed=True)
    directed = scipy.linalg.expm(-L1)[9][0]
    print("unconnected") if directed == 0 else print("connected, the minimum path is ",getMinPath(E1,100))

with open('Exercise1_Graph100b/graph2.pickle', 'rb') as f:
    E1 = pickle.load(f)
    L1 = getLaplacian(E1, 100, directed=True)
    directed = scipy.linalg.expm(-L1)[9][0]
    print("unconnected") if directed == 0 else print("connected, the minimum path is ",getMinPath(E1,100))
    
with open('Exercise1_Graph100b/graph3.pickle', 'rb') as f:
    E1 = pickle.load(f)
    L1 = getLaplacian(E1, 100, directed=True)
    directed = scipy.linalg.expm(-L1)[9][0]
    print("unconnected") if directed == 0 else print("connected, the minimum path is ",getMinPath(E1,100))
    
with open('Exercise1_Graph100b/graph4.pickle', 'rb') as f:
    E1 = pickle.load(f)
    L1 = getLaplacian(E1, 100, directed=True)
    directed = scipy.linalg.expm(-L1)[9][0]
    print("unconnected") if directed == 0 else print("connected, the minimum path is ",getMinPath(E1,100))
```

## Question 4

```python
import random 
# get random initial condition 
def simulate(x_0, N, T, L_list, switch_time, dt):
    # n node; T all simulation time
    time_arr = np.arange(0,T,dt)
    #switching order
    switch_arr = [0,1,2,3,0]
    
    x = np.zeros((N,len(time_arr)))
    x[:,0] = x_0
    
    switch_times = int(T/switch_time)
    period_times = int(switch_time/dt)

    for switch in range(0,switch_times):
        L = L_list[switch_arr[switch%5]]
        for i in range(period_times):
            if switch*period_times+i+1 == int(T/dt):
                continue
            x[:,switch*period_times+i+1] = (-1)*(np.dot(L,x[:,switch*period_times+i]))*dt + x[:,switch*period_times+i]
    
    return x
    
def plot(N, L, T=20):
    # plot the simulation curve, the initial condition is randomly distributed from [-10,10]
    x_0 = [random.uniform(-10,10) for _ in range(N)]
    
    time_arr = np.arange(0,T,0.001)
    
    x_01 = simulate(x_0, N, T, L, 0.1, 0.001)
    x_1 = simulate(x_0, N, T, L, 1, 0.001)

    plt.clf()
    #plot the 0.1 second switching one on the top
    for j in range(len(x_01)-1):
        plt.subplot(211)
        plt.plot(time_arr,x_01[j,:])
    
    for j in range(len(x_1)-1):
        plt.subplot(212)
        plt.plot(time_arr,x_1[j,:])
        
    plt.show()
```
